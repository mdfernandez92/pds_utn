# 2. Resolución
## 2.1. Inicialización y lectura de los datos ECG

```python
# Importamos bibliotecas a utilizar
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import scipy.io as sio
import scipy.signal as sig

# Configuramos tamaño de gráficos
mpl.rcParams['figure.figsize'] = (14,7)

# Configuramos tamaño de fuentes de gráficos
plt.rcParams.update({'font.size': 12})

# Leemos el dataset desde el archivo ECG.mat
mat_struct = sio.loadmat("ecg.mat")

# Serie temporal con el registro del ECG
xx = mat_struct["ecg_lead"]

# Definimos la frecuencia de sampling de la señal
fs = 1000

# Calculamos frecuencia de Nyquist 
fnyq = 0.5 * fs

# Para evitar aplicar logaritmo a amplitudes iguales a 0, vamos a sumar resolución de flotante a cada amplitud
eps = np.finfo(float).eps
```

## 2.2. Diseño de los filtros a partir de plantilla
## 2.2.1. Filtros IIR
En la siguiente sección se diseñan los siguientes filtros IIR:
* Filtro IIR Pasa-Banda combinando un Pasa-Bajo Butter y un Pasa-Alto Ellip
* Filtro IIR Pasa-Banda Butter
* Filtro IIR Pasa-Banda Ellip

```python
# Definimos plantilla del filtro
fs0 = 0.1
fc0 = 1
fc1 = 35
fs1 = 45
alfa_min = 60  # dB
alfa_max = 0.5 # dB

#-------------------------------------
# Filtro Pasa-Banda Butter
#-------------------------------------

# Definimos bandas de paso y corte (frec. normalizada a Nyquist)
wp = np.array([fc0, fc1]) / fnyq
ws = np.array([fs0, fs1]) / fnyq

# Diseñamos el filtro
sos_iir_bp_butter = sig.iirdesign(wp, ws, alfa_max, alfa_min/2, analog=False, ftype='butter', output='sos')

#-------------------------------------
# Filtro Pasa-Banda Butter
#-------------------------------------

# Definimos bandas de paso y corte (frec. normalizada a Nyquist)
wp = np.array([fc0, fc1]) / fnyq
ws = np.array([fs0, fs1]) / fnyq

# Diseñamos el filtro
sos_iir_bp_ellip = sig.iirdesign(wp, ws, alfa_max, alfa_min/2, analog=False, ftype='ellip',  output='sos')

#-------------------------------------------
# Filtro Pasa-Bajo Butter
#-------------------------------------------

# Definimos bandas de paso y corte (frec. normalizada a Nyquist)
wp = fc0 / fnyq
ws = 0.001 / fnyq

# Diseñamos el filtro
ba_lp_butter = sig.iirdesign(wp, ws, alfa_max, alfa_min/2, analog=False, ftype='butter', output='ba')

#-------------------------------------------
# Filtro Pasa-Alto Ellip
#-------------------------------------------

# Definimos bandas de paso y corte (frec. normalizada a Nyquist)
wp = fc1 / fnyq
ws = fs1 / fnyq

# Diseñamos el filtro
ba_hp_ellip = sig.iirdesign(wp, ws, alfa_max, alfa_min/2, analog=False, ftype='ellip', output='ba')

#--------------------------------------------------------
# Filtro Pasa-Banda = Pasa-Bajo Butter + Pasa-Alto Ellip
#--------------------------------------------------------

# Combinamos los filtros Pasa-Bajo y Pasa-Alto para formar un Pasa-Banda
sos_iir_bp_butter_ellip = sig.tf2sos(np.polymul(ba_lp_butter[0],ba_hp_ellip[0]),np.polymul(ba_lp_butter[1],ba_hp_ellip[1]))
```

## 2.2.2. Filtros FIR
En la siguiente sección se diseñan los siguientes filtros FIR:
* Filtro FIR Pasa-Banda combinando un Pasa-Bajo Remez y un Pasa-Alto Remez
* Filtro FIR Pasa-Banda combinando un Pasa-Bajo y Pasa-Alto Least Square

```python
#-------------------------------------------
# Filtro Pasa-Bajo Remez
#-------------------------------------------

# Definimos la cantidad de coeficientes
numtaps = 1501

# Definimos bandas de paso y corte
edges = np.array([0, fs0, fc0, fnyq])

# Definimos las ganancias en dB
gains = np.array([-200,0])

# Convertimos las ganancias a veces
gains = 10**(gains/20)

# Diseñamos el filtro
num_lp_remez = sig.remez(numtaps, edges, gains,fs=fs)

#-------------------------------------------
# Filtro Pasa-Alto Remez
#-------------------------------------------

# Definimos la cantidad de coeficientes
numtaps = 500

# Definimos bandas de paso y corte
edges = np.array([0, fc1, fs1, fnyq])

# Definimos las ganancias en dB
gains = np.array([0,-200])

# Convertimos las ganancias a veces
gains = 10**(gains/20)

# Diseñamos el filtro
num_hp_remez = sig.remez(numtaps, edges, gains,fs=fs)

#--------------------------------------------------------
# Filtro Pasa-Banda = Pasa-Bajo Remez + Pasa-Alto Remez
#--------------------------------------------------------

# Combinamos los filtros Pasa-Bajo y Pasa-Alto para formar un Pasa-Banda
num_fir_bp_remez = np.polymul(num_lp_remez,num_hp_remez)

#-------------------------------------------
# Filtro Pasa-Bajo Least Square
#-------------------------------------------

# Definimos la cantidad de coeficientes
numtaps = 1501

# Definimos bandas de paso y corte
edges = np.array([0, fs0, fc0, fnyq])

# Definimos las ganancias en dB
gains = np.array([-200,-(alfa_min),0,0])

# Convertimos las ganancias a veces
gains = 10**(gains/20)

# Diseñamos el filtro
num_lp_ls = sig.firls(numtaps, edges, gains,fs=fs)

#-------------------------------------------
# Filtro Pasa-Alto Least Square
#-------------------------------------------

# Definimos la cantidad de coeficientes
numtaps = 501

# Definimos bandas de paso y corte
edges = np.array([0, fc1, fs1, fnyq])

# Definimos las ganancias en dB
gains = np.array([0, 0, -(alfa_min), -200])

# Convertimos las ganancias a veces
gains = 10**(gains/20)

# Diseñamos el filtro
num_hp_ls = sig.firls(numtaps, edges, gains,fs=fs)

#--------------------------------------------------------
# Filtro Pasa-Banda = Pasa-Bajo LS + Pasa-Alto LS
#--------------------------------------------------------

# Combinamos los filtros Pasa-Bajo y Pasa-Alto para formar un Pasa-Banda
num_fir_bp_ls = np.polymul(num_lp_ls,num_hp_ls)
```

## 2.2.3. Respuesta en frecuencia de los filtros
En la siguiente sección se grafica la respuesta en frecuencia de los filtros diseñados

```python
#--------------------------------------------------------
# Graficación de la respuestas en frecuencia
#--------------------------------------------------------

# Obtenemos las respuestas en frecuencia de los filtros
w, h_iir_bp_butter       = sig.sosfreqz(sos_iir_bp_butter,       fs=fs,worN=np.logspace(-2,2,2048))
w, h_iir_bp_ellip        = sig.sosfreqz(sos_iir_bp_ellip,        fs=fs,worN=w)
w, h_iir_bp_butter_ellip = sig.sosfreqz(sos_iir_bp_butter_ellip, fs=fs,worN=w)

w, h_fir_bp_remez = sig.freqz(num_fir_bp_remez, fs=fs,worN=w)
w, h_fir_bp_ls    = sig.freqz(num_fir_bp_ls,    fs=fs,worN=w)

plt.figure()
plt.title("Espectros de amplitud - Filtros IIR y FIR")
plt.plot(w,20*np.log10(np.abs(h_iir_bp_butter_ellip)+eps), label="IIR - Butter+Ellip ({:d})".format(sos_iir_bp_butter_ellip.shape[0] * 2))
plt.plot(w,20*np.log10(np.abs(h_iir_bp_ellip)+eps), label="IIR - Ellip ({:d})".format(sos_iir_bp_ellip.shape[0] * 2))
plt.plot(w,20*np.log10(np.abs(h_iir_bp_butter)+eps),label="IIR - Butter ({:d})".format(sos_iir_bp_butter.shape[0] * 2))
plt.plot(w,20*np.log10(np.abs(h_fir_bp_ls)+eps),label="FIR - Least Square ({:d})".format(num_fir_bp_ls.shape[0]))
plt.plot(w,20*np.log10(np.abs(h_fir_bp_remez)+eps),label="FIR - Remez ({:d})".format(num_fir_bp_remez.shape[0]))
plt.ylabel("|H(f)| (dB)")
plt.xlabel("Frecuencia (Hz)")
plt.ylim(-130,10)
plt.grid()
plt.legend()
plt.show()
```

## 2.3. Aplicación de los filtros a la señal ECG
En la siguiente sección se aplicaran los filtros diseñados en instantes de la señal ECG donde se tiene ruido interferente y donde no se lo tiene, para observar el comportamiento de los mismos.

```python
# Obtenemos las salidas al aplicar los diversos filtros a la señal de entrada
yy_iir_bp_butter       = sig.sosfiltfilt(sos_iir_bp_butter,      xx,axis=0)
yy_iir_bp_ellip        = sig.sosfiltfilt(sos_iir_bp_ellip,       xx,axis=0)
yy_iir_bp_butter_ellip = sig.sosfiltfilt(sos_iir_bp_butter_ellip,xx,axis=0)
yy_fir_bp_remez        = sig.filtfilt(num_fir_bp_remez, 1,       xx,axis=0)
yy_fir_bp_ls           = sig.filtfilt(num_fir_bp_ls,    1,       xx,axis=0)

# Definimos indexadores para señales con ruido
idx_interf_low_freq = ( 
    np.array([12, 12.4]) *60*fs, # minutos a muestras
    np.array([15, 15.2]) *60*fs, # minutos a muestras
)

# Definimos indexadores para señales sin ruido
idx_no_interf = ( 
    np.array([5, 5.2]) *60*fs, # minutos a muestras
    [4000, 5500], # muestras
    [10e3, 11e3], # muestras
)

#---------------------------------------------------------
# Señales ECG con interferencia
#---------------------------------------------------------

for i in range(0,len(idx_interf_low_freq)):
    # Indexamos los valores de entrada de la señal con inteferencia
    xx_i = xx[int(idx_interf_low_freq[i][0]):int(idx_interf_low_freq[i][1])]
    
    # Indexamos los valores de salida tras aplicar los filtros
    yy_i_iir_bp_butter       = yy_iir_bp_butter      [int(idx_interf_low_freq[i][0]):int(idx_interf_low_freq[i][1])]
    yy_i_iir_bp_ellip        = yy_iir_bp_ellip       [int(idx_interf_low_freq[i][0]):int(idx_interf_low_freq[i][1])]
    yy_i_iir_bp_butter_ellip = yy_iir_bp_butter_ellip[int(idx_interf_low_freq[i][0]):int(idx_interf_low_freq[i][1])]
    yy_i_fir_bp_remez        = yy_fir_bp_remez       [int(idx_interf_low_freq[i][0]):int(idx_interf_low_freq[i][1])]
    yy_i_fir_bp_ls           = yy_fir_bp_ls          [int(idx_interf_low_freq[i][0]):int(idx_interf_low_freq[i][1])]

    # Graficamos la entrada y la salida obtenida tras aplicar cada filtro
    plt.figure()
    plt.title("Señal ECG con ruido de baja frecuencia")
    plt.plot(xx_i,label="original")
    plt.plot(yy_i_iir_bp_butter_ellip,label="IIR - Butter+Ellip")
    plt.plot(yy_i_iir_bp_ellip,       label="IIR - Ellip")
    plt.plot(yy_i_iir_bp_butter,      label="IIR - Butter")
    plt.plot(yy_i_fir_bp_remez,       label="FIR - Remez")
    plt.plot(yy_i_fir_bp_ls,          label="FIR - Least Square")
    plt.ylabel("Amplitud")
    plt.xlabel("Número de muestra")
    plt.legend()
    plt.show()
    
#---------------------------------------------------------
# Señales ECG sin interferencia
#---------------------------------------------------------

for i in range(0,len(idx_no_interf)):
    # Indexamos los valores de entrada de la señal con inteferencia
    xx_i = xx[int(idx_no_interf[i][0]):int(idx_no_interf[i][1])]
    
    # Indexamos los valores de salida tras aplicar los filtros
    yy_i_iir_bp_butter       = yy_iir_bp_butter      [int(idx_no_interf[i][0]):int(idx_no_interf[i][1])]
    yy_i_iir_bp_ellip        = yy_iir_bp_ellip       [int(idx_no_interf[i][0]):int(idx_no_interf[i][1])]
    yy_i_iir_bp_butter_ellip = yy_iir_bp_butter_ellip[int(idx_no_interf[i][0]):int(idx_no_interf[i][1])]
    yy_i_fir_bp_remez        = yy_fir_bp_remez       [int(idx_no_interf[i][0]):int(idx_no_interf[i][1])]
    yy_i_fir_bp_ls           = yy_fir_bp_ls          [int(idx_no_interf[i][0]):int(idx_no_interf[i][1])]

    # Graficamos la entrada y la salida obtenida tras aplicar cada filtro
    plt.figure()
    plt.title("Señal ECG sin ruido")
    plt.plot(xx_i,label="original")
    plt.plot(yy_i_iir_bp_butter_ellip,label="IIR - Butter+Ellip")
    plt.plot(yy_i_iir_bp_ellip,       label="IIR - Ellip")
    plt.plot(yy_i_iir_bp_butter,      label="IIR - Butter")
    plt.plot(yy_i_fir_bp_remez,       label="FIR - Remez")
    plt.plot(yy_i_fir_bp_ls,          label="FIR - Least Square")
    plt.ylabel("Amplitud")
    plt.xlabel("Número de muestra")
    plt.legend()
    plt.show()
```

* Combinar filtros es adecuado para cumplir con plantillas de diseño exigentes y asimétricas. A su vez permite obtener características del filtro que no son posibles de obtener con un solo filtro (como por ejemplo dar más importancia a una banda que a otra).

* El filtro Butter a diferencia del Butter+Ellip cumplió muy bien con la plantilla en baja frecuencia. De todas formas, tener en cuenta el Butter+Ellip es de mucho menos orden.

* El filtro Remez es más eficiente respecto al costo computacional que el del filtro Least Square porque para un mismo orden logro una especificación mas exigente en alta frecuencia. Podríamos sacarle orden.

* El filtro Ellip cumplió con los 30db en 0.1Hz con una respuesta equi-ripple (ningún máximo excede los 30db) lo cual es muy bueno ya que con un orden mucho menor cumplió la especificación de forma aceptable en comparación con el filtro Butter. De todas formas, la respuesta de fase de este filtro es dañina.

Respecto a la salida obtenida tras aplicar los filtros a la señal ECG:

* Todos los filtros respondieron bien al ruido de baja frecuencia: lograron quitar la oscilación de muy baja frecuencia que se observa montada en las señales con ruido y en las señales sin ruido se quita el valor medio.

* Todos los filtros logran suavizar un poco la señal pero exhiben una mala respuesta temporal frente a impulsos. Esto puede observarse ante la presencia del complejo QRS: tras ocurrir el impulso aparecen en las cercanías del evento pequeñas oscilaciones de alta frecuencia que no existían en la señal original. Cuando definimos filtros con flancos de transición muy abruptos arruinamos la respuesta temporal del mismo. Este efecto temporal no tenemos forma de controlarlo en el dominio de la frecuencia.

* Se observa en algunos casos disminución del nivel en los latidos ventriculares respecto a la señal original

Como conclusión final podemos decir que los filtros aplicados en señales con ruido lograron filtrar adecuadamente la interferencia, pero en señales no contaminadas no son 100% inocuos.
